```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_list = [1,1,1,1,1,1]
```

```python
t1 = tf.constant(test_list)
print(t1)
```

```python
t2 = tf.zeros(shape = (100,)) 
print(type(t2))
print(t2)
```

```python
PI = np.pi
t3 = tf.ones(shape = (100,))*PI
print(t3)
```

```python
test_list = [[1,2,3],[4,5,6]]
t1 = tf.Variable(test_list)
print(t1)

#t2 = tf.ones(shape = t1.shape)
t2 = tf.ones_like(t1)  # t1 과 같은 shape의 tensor를 생성
print(t2)
```

```python
tf.random.set_seed(0) # 일정한 random seed 넣어주기

t1 = tf.random.normal(shape = (10,10))

# 이거는 확실하게 알아두기,  다음 들어올 tesnor의 형태를 조정해주는데 사용해주기도 함
print("t1.shape", t1.shape)
print("t1.dtype", t1.dtype)

t1 = tf.constant(test_list, dtype = tf.float32) # data type 정해줘서 tensor 만들어 주기
print(t1.dtype)
```

```python
t2 = tf.random.normal(mean = 3, stddev = 1, shape = (1000,))
# print(t2)
fig , ax = plt.subplots(figsize = (15,15))
ax.hist(t2.numpy(), bins = 30)
ax.tick_params(labelsize = 20)
```

```python
t2 = tf.random.uniform(shape = (100000,), minval = -10, maxval = 10)
fig , ax = plt.subplots(figsize = (15,15))
ax.hist(t2.numpy(), bins = 30)
ax.tick_params(labelsize = 20)
```

```python
t2  = tf.random.poisson(shape = (10000,), lam = 5)
fig , ax = plt.subplots(figsize = (15,15))
ax.hist(t2.numpy(), bins = 30)
ax.tick_params(labelsize = 20)
```
